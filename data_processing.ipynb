```python
import re

def remove_tashkeel(text):
    tashkeel_pattern = r'[\u064B-\u065F]'  
    return re.sub(tashkeel_pattern, '', text)

ayah = "مَا تَتْلُو الشَّيَاطِينُ عَلَىٰ مُلْكِ سُلَيْمَانَ"
print(remove_tashkeel(ayah)) 
```

```python
ayah_table["ayah_with_out_tashkeel"]=ayah_table["ayah_text"].apply(lambda x:remove_tashkeel(x))
ayah_table
```

#apply same concept by using sql 

DELIMITER $$

CREATE FUNCTION remove_tashkeel(textvalue VARCHAR(255)) RETURNS VARCHAR(255)
BEGIN
    SET textvalue = REPLACE(textvalue, 'َ', '');
    SET textvalue = REPLACE(textvalue, 'ً', '');
    SET textvalue = REPLACE(textvalue, 'ُ', '');
    SET textvalue = REPLACE(textvalue, 'ٌ', '');
    SET textvalue = REPLACE(textvalue, 'ِ', '');
    SET textvalue = REPLACE(textvalue, 'ٍ', '');
    SET textvalue = REPLACE(textvalue, 'ْ', '');
    SET textvalue = REPLACE(textvalue, 'ّ', '');
    RETURN textvalue;
END $$

DELIMITER ;
